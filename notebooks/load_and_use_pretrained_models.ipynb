```python
import gensim
import os


# model path
model_save_dir = 'models/pretrained/2'
model_name = "word2vec-from-posts.bin"
model_path = os.path.join(model_save_dir, model_name)

# load word2vec model
try:
    loaded_word2vec_model = gensim.models.Word2Vec.load(model_path)
    print(f"Word2Vec model loaded successfully from: {model_path}")
except Exception as e:
    print(f"Error loading Word2Vec model: {e}")
    loaded_word2vec_model = None

if loaded_word2vec_model:
    test_word = input("Enter the word to text : ")

    # check if exists in word vocabulary
    if test_word in loaded_word2vec_model.wv:
        print(f"\nWords similar to '{test_word}':")
        for word, similarity in loaded_word2vec_model.wv.most_similar(test_word, topn=10):
            print(f"  {word}: {similarity:.4f}")
    else:
        print(
            f"\n'{test_word}' not found in the loaded model's vocabulary. Try another word.")
else:
    print("Cannot demonstrate similar word prediction as the model could not be loaded.")


# function to use the previous one to predict/complete a sentence
def complete_sentence(model, sentence, max_words=5):
    """
    Complete a sentence by predicting next words
    using Word2Vec similarity.
    """

    words = sentence.split()

    for _ in range(max_words):

        last_word = words[-1]

        if last_word not in model.wv:
            print(f"'{last_word}' not in vocabulary. Stopping.")
            break

        # get most similar word
        similar_words = model.wv.most_similar(last_word, topn=10)

        
        for word, score in similar_words:
            if word not in words:
                next_word = word
                break

        words.append(next_word)

    return " ".join(words)
```
