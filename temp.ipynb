# Word2Vec — Moroccan Dialect
Train the model, generate embeddings, and predict next words.

```python
from model import train, embed_dataset, predict_next_word
```

## Step 1 — Train
Read `data/global.txt` and train the Skip-gram model. Saves weights to `model/`.

```python
train(
    data_path = "data/global.txt",
    save_dir  = "model",
    epochs    = 10
)
```

## Step 2 — Generate Word Embeddings
Convert every sentence in the dataset to a vector. Saves result to `data/embedding.csv`.

```python
embed_dataset(
    data_path   = "data/global.txt",
    output_path = "data/embedding.csv",
    save_dir    = "model"
)
```

Preview the first few rows of the embedding CSV.

```python
import pandas as pd

df = pd.read_csv("data/embedding.csv")
print(f"Shape: {df.shape}")
df.head()
```

## Step 3 — Predict Next Words
Give a sentence and get the most likely next words back.

```python
results = predict_next_word(
    input_sentence = "ana machi",
    save_dir       = "model",
    length         = 5        # change this to get more or fewer predictions
)

for word, score in results:
    print(f"{word:<20} {score:.4f}")
```
