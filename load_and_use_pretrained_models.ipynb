```python
import gensim
import os

# print(os.listdir())



# Define the path where the model was saved
model_save_dir = 'models/pretrained/'
model_path = os.path.join(model_save_dir, "word2vec-from-posts.bin")

# 1. Load the Word2Vec model
try:
    loaded_word2vec_model = gensim.models.Word2Vec.load(model_path)
    print(f"Word2Vec model loaded successfully from: {model_path}")
except Exception as e:
    print(f"Error loading Word2Vec model: {e}")
    loaded_word2vec_model = None

if loaded_word2vec_model:
    # 2. Define a test word
    test_word = "الاسلام"  # You can change this word to experiment

    # 3. Check if the test word exists in the loaded model's vocabulary
    if test_word in loaded_word2vec_model.wv:
        # 4. Use most_similar() to find and print top N similar words
        print(f"\nWords similar to '{test_word}':")
        for word, similarity in loaded_word2vec_model.wv.most_similar(test_word, topn=10):
            print(f"  {word}: {similarity:.4f}")
    else:
        print(
            f"\n'{test_word}' not found in the loaded model's vocabulary. Try another word.")
else:
    print("Cannot demonstrate similar word prediction as the model could not be loaded.")
```
